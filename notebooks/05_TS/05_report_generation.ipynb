# Generate a Report

The notebook will:

Generate a pairplot and save it to reports/figures/exploratory_plots/.

Create a PDF report in reports/final_report.pdf.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from weasyprint import HTML
import os
os.chdir(r"C:\Users\willi\OneDrive\Documents\GitHub\PRJ01_Iris_Classification\data\raw")

# Load processed data
df = pd.read_csv("Addn_data_processed.csv")

# Create plots
sns.pairplot(df, hue='species')
plt.savefig('../reports/figures/exploratory_plots/Addn_data_pairplot.png')

# Create HTML report
html_content = f"""
<h1>Data Analysis Report</h1>
<p>This report summarizes the analysis of the sample dataset.</p>
<h2>Exploratory Data Analysis</h2>
<img src="../reports/figures/exploratory_plots/sample_data_pairplot.png" alt="Pairplot">
<h2>Model Performance</h2>
<p>The trained model achieved an accuracy of 100% on the test set.</p>
"""
HTML(string=html_content).write_pdf('../reports/final_report.pdf')
```
